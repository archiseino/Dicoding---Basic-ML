### Handling the Missing Value
As we can see, we have 1460 rows of data and some of the value has a missing value, so the first thing we can do is to seperate the data has few and more missing value

```python
## Seperating the dataset train less and over 1000 of missing value
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values >= 1000].index
```

### Filling the Missing Value
For filling the missing value, firstly  we need to seperate the type of data for the purpose of the filling.

```python
# Contoh mengisi nilai yang hilang dengan median untuk kolom numerik
numeric_features = train_dataset[less].select_dtypes(include=['number']).columns
train_dataset[numeric_features] = train_dataset[numeric_features].fillna(train_dataset[numeric_features].median())
```

### Notes:
For number, the way of filling the missing value is by fill it with the median value of that particular column.

For string and object, this get interesting, we have two ways of filling it in.
- Using modus (most show up data) since it's the better representation in categorical data.
- Fill it using the new Category (such as `Unknown` or `Missing`)

```python
### We fill it with the modus data for the missing value of string and object
categorical_features = train_dataset[less].select_dtypes(include=["object"]).columns

## Filling the missing value for each column
for column in categorical_features:
    train_dataset[column] = train_dataset[column].fillna(train_dataset[column].mode()[0])
```

If the column with many missing values, we can consider it to remove the column (if the limit is above 75% as the best practice)

```python
## Remove the column with many missing values
df = train_dataset.drop(columns=over)
```

### Doing verification of the new data after filetering the missing value

### Handling Outliers
Outlier is a blocker in a machine learning when there's some error / problem on the data itself.

One way to handle this is using the IQR Method (Interquartile Range). IQR is a distance between the first (Q1) and third (Q3) quartile in a sorted data. 

Value outside of this is a outlier.

```python
### Checking if the dataset has outlier
import seaborn as sns
import matplotlib.pyplot as plt

# for feature in numeric_features:
#     plt.figure(figsize=(10,6))
#     sns.boxplot(x=df[feature])
#     plt.title(f"Box Plot of {feature}")
#     plt.show()
```

### Detection and Removing the Outlier
We can do several things for this, but firstly we can calcualte the IQR, Q1 and Q3
- Q1 is the value of the 25th (1/4) value data
- Q3 is the value of the 75th (3/4) value data
- IQR is the distance between Q3 and Q1

Lastly we can calculate the Bottom and Upper limit of this
- Bottom limit = Q1 - 1.5 * IQR
- Upper limit = Q3 + 1.5 * IQR

```python
## Simple method for detcting the outliers using IQR
Q1 = df[numeric_features].quantile(0.25)
Q3 = df[numeric_features].quantile(0.75)
IQR = Q3 - Q1
```

```python
## Delete the outliers 
condition = ~((df[numeric_features] < (Q1 - 1.5 * IQR)) | (df[numeric_features] > (Q3 + 1.5 * IQR))).any(axis=1)
df_filetered_numeric = df.loc[condition, numeric_features]

## Combining the categorical value
categorical_features = df.select_dtypes(include=["object"]).columns
df = pd.concat([df_filetered_numeric, df.loc[condition, categorical_features]])
```

### Normalization and Standarization Data

Since the data we have for numeric part has different scale, We need to do the normalization / standarication, but here's the diff.

![image.png](https://dicoding-web-img.sgp1.cdn.digitaloceanspaces.com/original/academy/dos-a3d76eeb567571fdb2f8d40f95b0dff720241015110131.jpeg)

- Normalization (known as the min-max scaling) is to change the feature in some interval (from 0 and 1) usually.
- Standarization (known as the Z-score scaling) is to change the scale of a feature to have the mean of 0 and standard deviation of 1 of each column (Or we can called it Normal Distribution)

So in the nutshell
- Normalization: transforms data into the range [0, 1]. Ideal for distance-based models and non-normally distributed data.
- Standardization: changing the data so that the mean is or approaches 0 and the standard deviation is 1. This helps models that require a normal distribution or data with different scales to be more ideal in the training process.

```python
## Standarization example
from sklearn.preprocessing import StandardScaler
 
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

To handle convertion from the categorical and numeric, we can use something that called the Encoding. There's some examples of them
- One-Hot Encoding: Changing the Every category into binary column with (0 and 1)
- Label Encoding: Giving a label for each category, even though it only works for small set of category data
- Ordinal Encoding: Mapping Convertion from each category into numeric value.

### Making Correlation Matrix
Matrix Correlation shows the important relation between variable in a dataset.

```python
# Visualisasi korelasi antar variabel numerik
plt.figure(figsize=(12, 10))
correlation_matrix = df_lencoder.corr()
 
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

#### Notes:
- Korelasi Positif: nilai korelasi positif menunjukkan bahwa saat satu variabel meningkat, variabel lain juga cenderung meningkat. Nilai korelasi +1 menunjukkan korelasi positif sempurna.
- Korelasi Negatif: nilai korelasi negatif menunjukkan bahwa saat satu variabel meningkat, variabel lain cenderung menurun. Nilai korelasi -1 menunjukkan korelasi negatif sempurna.
- Korelasi Nol: nilai korelasi 0 menunjukkan bahwa tidak ada hubungan linier antara dua variabel.

You even can break this even further to see how does one variable corellate with each other.

```python
# Menghitung korelasi antara variabel target dan semua variabel lainnya
target_corr = df_lencoder.corr()['SalePrice']
 
# (Opsional) Mengurutkan hasil korelasi berdasarkan korelasi
target_corr_sorted = target_corr.abs().sort_values(ascending=False)
 
plt.figure(figsize=(10, 6))
target_corr_sorted.plot(kind='bar')
plt.title(f'Correlation with SalePrice')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.show()
```
